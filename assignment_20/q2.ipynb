### LABLE ENCODING


```python
import pandas as pd 
```

```python
df = pd.read_csv("Used_Bikes.csv")
```

```python
df.head()
```

```python
df = df.drop("city",axis=1)
```

```python
df
```

```python
df["owner"].value_counts()
```

### this is lable encoding where we assign a particular value to a particular data

```python
dt = {"First Owner":1,"Second Owner":2,"Third Owner":3,"Fourth Owner Or More":4}
```

```python
df["owner"] = df["owner"].map(dt)
```

```python
brands = list(df["brand"].unique())
brands
```

```python
dt={}
for i,items in enumerate(brands):
    dt[items]=i
dt
```

### this is also label encoding where each brand gets a different value

#### disavantage of it is that model gives higher priority to the the higher value

```python
df["brand"] = df["brand"].map(dt)
```

```python
df
```

## ONE HOT ENCODING

```python
df.head()
```
